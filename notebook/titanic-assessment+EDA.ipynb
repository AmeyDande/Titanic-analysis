# About this Dataset

RMS Titanic was a British luxury ocean liner, renowned as one of the largest and most opulent ships of its era. Built by Harland & Wolff in Belfast, it measured about 883 feet long, 92 feet wide, and could carry over 3,500 passengers and crew
On April 10, 1912, Titanic departed Southampton for New York with 2,208 aboard. It struck an iceberg on April 14, sinking early on April 15 and claiming around 1,500 lives—one of history's deadliest peacetime maritime disasters


This data shows the limited amount of passengers which were there on, **Titanic** that day

### Column description


1. ```Survived``` :- It shows wether that passenger survived or not using *0 = No, 1 = Yes*
2. ```Pclass``` :- Ticket class	*1 = 1st;Upper, 2 = 2nd;Middle, 3 = 3rd;Lower*
3. ```Name``` :- Shows the name of the passenger 
4. ```Sex``` :-  Shows the gender of tha passenger
5. ```Age``` :- Shows the age of the passeger *Age is fractional if less than 1. If the age is estimated, is it in the form of xx.5*
6. ```SibSp``` :- The dataset defines family relations in this way 
; Sibling = brother, sister, stepbrother, stepsister
; Spouse = husband, wife *(mistresses and fiancés were ignored)*
7. ```Parch``` :- The dataset defines family relations in this way ;
Parent = mother, father ;
Child = daughter, son, stepdaughter, stepson ;
*Some children travelled only with a nanny, therefore parch=0 for them.*
8. ```Ticket``` :- Shows the ticket number of the passenger 
9. ```Fare``` :- Shows the fare of the passenger 
10. ```Cabin``` :- Shows the Cabin number of the passenger 
11. ```Embarked``` :- Port of Embarkation *C = Cherbourg, Q = Queenstown, S = Southampton*


#### Issues with the dataset

1. Dirty Data

- There are some missing values in ```Age``` column
- There are too many missing values in ```Cabin``` column
- very few values of ```Embarked``` are missing

2. Messy data


Thers no messy data as such

#### Define 
- Filling missing values of columns and ```Embarked``` using and mode() 
- Dropping column ```Cabin``` cause it has too many missing values
- Creating ```lone_traveler``` column to identify which passenger is travelling alone 


````python
df.isna().sum()
````

````python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
````

````python
df.drop(columns=['Cabin', 'PassengerId'], inplace=True)
````

````python
df.isna().sum()
````

````python
df['lone_traveler'] = df['SibSp'] + df['Parch'] == 0 
````

````python
df['lone_traveler'] = df['lone_traveler'].replace({True: 'Yes', False: 'No'})
````
